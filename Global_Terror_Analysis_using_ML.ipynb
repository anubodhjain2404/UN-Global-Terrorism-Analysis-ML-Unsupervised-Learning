# **Objective**

To detect geographical terrorism hotspots without predefined cluster count.

This means we want to identify high-risk regions automatically from the data, without telling the model how many clusters (hotspots) to form.

# **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA

pd.set_option('display.max_columns', None)
```

# **Selecting Features**

```python
data = df[['latitude', 'longitude', 'region_txt', 'attacktype1_txt']]

# Remove missing values
data = data.dropna()

# Remove duplicates
data = data.drop_duplicates()

print("Shape:", data.shape)
```

# **Data Scaling**

```python
data = data.sample(n=20000, random_state=42)
```

```python
data_encoded = pd.get_dummies(data, columns=['region_txt', 'attacktype1_txt'])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_encoded)
```

# **1) K-Means Clustering**

```python
wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('K')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
data['KMeans_Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
sns.scatterplot(x=data['longitude'], y=data['latitude'], hue=data['KMeans_Cluster'])
plt.title('K-Means Clusters')
plt.show()
```

# **Insights:**
* Data grouped into clusters based on similarity
* Some clusters show high activity regions
* Requires predefined K
* Cannot detect irregular shapes

# **2) PCA(Dimensionality Reduction)**

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

print(pca.explained_variance_ratio_)
```

```python
plt.scatter(pca_data[:,0], pca_data[:,1], alpha=0.5)
plt.title('PCA Visualization')
plt.show()
```

# **Insights:**
* Data reduced to 2 components
* Helps visualize patterns clearly
* Shows separation between clusters
* Not used for clustering directly

# **3) DBSCAN**

```python
dbscan = DBSCAN(eps=0.8, min_samples=10)

data['DBSCAN_Cluster'] = dbscan.fit_predict(scaled_data)

print(np.unique(data['DBSCAN_Cluster']))
```

```python
print(data['DBSCAN_Cluster'].value_counts())
```

```python
sns.scatterplot(x=data['longitude'], y=data['latitude'], hue=data['DBSCAN_Cluster'])
plt.title('DBSCAN Hotspots')
plt.show()
```

```python
hotspots = data[data['DBSCAN_Cluster'] != -1]
noise = data[data['DBSCAN_Cluster'] == -1]

print("Hotspots:", len(hotspots))
print("Noise:", len(noise))
```

```python
plt.figure(figsize=(14,8))

sns.scatterplot(
    data=hotspots,
    x='longitude',
    y='latitude',
    hue='DBSCAN_Cluster',
    palette='tab20',
    s=40
)

plt.scatter(
    noise['longitude'],
    noise['latitude'],
    color='black',
    s=10,
    label='Noise'
)

plt.title('DBSCAN Hotspots with Noise')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.show()
```

# **Visualization Outcomes**

* The visualization shows that terrorism incidents are concentrated in specific dense regions (hotspots)
* DBSCAN successfully separates these hotspots from isolated events (noise)
* The clusters are irregular in shape, reflecting real-world geographical patterns
* It also highlights variation in intensity across regions, with some areas having significantly higher activity

# **Insights:**
